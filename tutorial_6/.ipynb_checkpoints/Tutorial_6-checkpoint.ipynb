### Part_C:-

```python
def avg_count(Data):
    male_count=0
    male_age=0
    male_Spending_score=0
    female_count=0
    female_age=0
    female_Spending_score=0
    for i in range (len(Data['Gender'])):
        if Data.iloc[i]['Gender']=='Male':
            male_count=male_count+1
            male_age=male_age+Data.iloc[i]['Age']
            male_Spending_score=male_Spending_score+Data.iloc[i]['Spending Score (1-100)']
        if Data.iloc[i]['Gender']=='Female':
            female_count=female_count+1
            female_age=female_age+Data.iloc[i]['Age']
            female_Spending_score=female_Spending_score+Data.iloc[i]['Spending Score (1-100)']
    Avg_male_age=male_age/male_count
    Avg_female_age=female_age/female_count
    Avg_male_Spending_score=male_Spending_score/male_count
    Avg_female_Spending_score=female_Spending_score/female_count
    return ({'Male':[Avg_male_age,Avg_male_Spending_score],
            'Female':[Avg_female_age,Avg_female_Spending_score]})
```

```python
avg_val=avg_count(Data)
avg_val
```

```python
plt.barh(avg_val['Male'][0],avg_val['Male'][1],label='Male')
plt.barh(avg_val['Female'][0],avg_val['Female'][1],label='Female')
plt.ylabel('Average Age')
plt.xlabel('Average Spending Score')
plt.legend()
plt.xlim(45,52)
plt.show()
```

```python
def avg_income_spending(Data):
    male_count=0
    male_income=0
    male_Spending_score=0
    female_count=0
    female_income=0
    female_Spending_score=0
    for i in range (len(Data['Gender'])):
        if Data.iloc[i]['Gender']=='Male':
            male_count=male_count+1
            male_income=male_income+Data.iloc[i]['Annual Income (k$)']
            male_Spending_score=male_Spending_score+Data.iloc[i]['Spending Score (1-100)']
        if Data.iloc[i]['Gender']=='Female':
            female_count=female_count+1
            female_income=female_income+Data.iloc[i]['Annual Income (k$)']
            female_Spending_score=female_Spending_score+Data.iloc[i]['Spending Score (1-100)']
    Avg_male_income=male_income/male_count
    Avg_female_income=female_income/female_count
    Avg_male_Spending_score=male_Spending_score/male_count
    Avg_female_Spending_score=female_Spending_score/female_count
    return ({'Male':[Avg_male_income,Avg_male_Spending_score],
            'Female':[Avg_female_income,Avg_female_Spending_score]})
```

## Part_E,F,G,H:-

```python
from sklearn.cluster import KMeans
```

```python
plt.scatter(Data['Annual Income (k$)'],Data['Spending Score (1-100)'])
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Scatter Plot')
plt.show()
```

```python
def Best_k(Data,k):
    inertia=[]
    for i in range(1,k):
        K_mean=KMeans(n_clusters=i)
        K_mean.fit(list(zip(Data['Annual Income (k$)'].values,Data['Spending Score (1-100)'].values)))
        inertia.append(K_mean.inertia_)
    plt.plot(range(1,k),inertia,marker='o')
    plt.title('Elbow Method')
    plt.xlabel('Number Of Clusters')
    plt.ylabel('WCSS')
```

```python
Best_k(Data,11)
```

```python
k=5
K_mean=KMeans(n_clusters=k)
K_mean.fit(list(zip(Data['Annual Income (k$)'].values,Data['Spending Score (1-100)'].values)))
plt.scatter(Data['Annual Income (k$)'],Data['Spending Score (1-100)'],c=K_mean.labels_)
for i in range(k):
    plt.scatter(K_mean.cluster_centers_[i][0],K_mean.cluster_centers_[i][1],c='Red',marker='*')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.title('Clutering Scatterplot')
plt.plot()
```

### Part_c,d,e,f:-

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
redu_data=pca.fit_transform(Data1)
```

```python
redu_data
redu_data=pd.DataFrame(redu_data)
```

```python
inertia=[]
k=11
for i in range(1,k):
    K_mean=KMeans(n_clusters=i)
    K_mean.fit(redu_data)
    inertia.append(K_mean.inertia_)
plt.plot(range(1,k),inertia,marker='o')
plt.title('Elbow Method')
plt.xlabel('Number Of Clusters')
plt.ylabel('WCSS')
plt.plot()
```

##### Best Cluster Value is 2

```python
k=2
K_mean=KMeans(n_clusters=k)
K_mean.fit(redu_data)
plt.scatter(redu_data[0],redu_data[1],c=K_mean.labels_)
for i in range(k):
    plt.scatter(K_mean.cluster_centers_[i][0],K_mean.cluster_centers_[i][1],c='Red',marker='*')
plt.xlabel('Dimention_1')
plt.ylabel('Dimention_2')
plt.title('Reduced Dimension Clustered Plot')
plt.plot()
```
